```python
from random import randint

def gen_random_list(L, R, n):

    int_list = []

    for k in range(n):
        r = randint(L, R)
        int_list.append(r)

    return int_list

if __name__ =='__main__':
    roll_dice_list = gen_random_list(1, 6, 10)
    print(roll_dice_list)
```

```python
def are_anagram(word1, word2):
    

    word1_sorted = sorted(word1)
    word2_sorted = sorted(word2)
    print(word1_sorted, word2_sorted)

    return word1_sorted==word2_sorted

if __name__=='__main__':
    two_words = input('Enter two space separated words: ')
    word1, word2, = two_words.split()

    if are_anagram(word1, word2):
        print(f'{word1}, {word2} : anagram')
    else:
        print(f'{word1}, {word2} : not anagram')
```

```python
import math

def std_dev(lst):

    n = len(lst)
    mean = sum(lst)/n
    summation = sum([(x-mean)**2 for x in lst])
    variance = summation/(n-1)

    std_dev = math.sqrt(variance)
    return mean, std_dev

if __name__=='__main__':
    lst = [3,4,5,6,6,7,1,2,3]
    avg, std = std_dev(lst)
    print(f'average: {avg:.2f}, standard deviation :{std:.5f}')
```
