```python
# 유니버설 함수
# 1. 단항 유니버설 함수 : 매개변수 하나만 갖는 함수
# 2. 이항 유니버설 함수 : 매개변수 두 개 갖는 함수수
```

```python
#삼각함수 (역삼각함수는 arcsin, arccos, arctan)
theta = np.linspace(0, np.pi, 3)
print("theta= ", theta)
print("sin(theta) = ", np.sin(theta))
print("cos(theta) = ", np.cos(theta))
print("tan(theta) = ", np.tan(theta))
```

```python
#지수함수//  로그함수는 ln(x), log2(x), log10(x)와 같이 표기
print("x = ", x)
print("e^x = ", np.exp(x))
print("2^x = ", np.exp2(x))
print("3^x = ", np.power(3,x))
```

```python
#out인자 : 지정한 배열에 직접 연산 결과 표기 가능
import numpy as np
x=np.arange(5)
y=np.zeros(10)
np.power(2,x,out=y[::2]) # 2씩 건너띈다.
print(y)
#대규모 연산의 경우 임시 배열을 생략하여 메모리와 연산시간 절감
```
